# Pie chart
A pie chart is a circular statistical graphic used to represent data in the form of proportions or percentages. It is divided into slices, with each slice representing a category or portion of the whole dataset. The size of each slice is proportional to the quantity or percentage it represents.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn_data.csv')
df
```

Here are the counts of different payment method

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
payment_method_counts = df['PaymentMethod'].value_counts()
payment_method_counts
```

There is no pie chart method bulid in seaborn. Version = 0.13.2

```python
plt.figure(figsize = (10,6))
plt.pie(x = payment_method_counts.values, labels = payment_method_counts.index)
plt.title('Payment Method Distribution')
plt.show()
```

Add percentage

```python
plt.figure(figsize = (10,6))
plt.pie(x = payment_method_counts.values, labels = payment_method_counts.index, autopct = '%.2f%%')
plt.title('Payment Method Distribution')
plt.show()
```

We could add annotations to show the percentage of each slice of the pie chart as well as explode a particular slice for emphasis:

```python
plt.figure(figsize = (5, 3))
plt.pie(
    x = payment_method_counts.values,
    labels = payment_method_counts.index,
    autopct = '%.1f%%',
    explode = [0, 0, 0.3, 0]
)
plt.title('Pie chart showing the distribution by payment method')
plt.show()
```
