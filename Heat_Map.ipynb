#Heat Map
A heat map is primarily used for analysing patterns, trends, and intensity within a dataset by representing values through colour gradients. It’s ideal for correlation analysis (e.g., relationships between variables), spatial analysis (e.g., density or distribution across regions), and performance tracking (e.g., identifying high- and low-performing areas in sales, marketing, or system monitoring).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn_data.csv')
df
```

```python
df.info()
```

```python
df.columns
```

Let's create correlations between the numeric variables in the dataset.

```python
correlation_matrix = df[['tenure','MonthlyCharges','TotalCharges']].corr()
correlation_matrix
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(data = correlation_matrix)
plt.title('Correlation Matrix Between tenure, MonthlyCharges,TotalCharges')
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(data = correlation_matrix, annot = True)
plt.title('Correlation Matrix Between tenure, MonthlyCharges,TotalCharges')
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f')
plt.title('Correlation Matrix Between tenure, MonthlyCharges,TotalCharges')
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f', linewidths = 0.5, cmap = 'coolwarm')
plt.title('Correlation Matrix Between tenure, MonthlyCharges,TotalCharges')
plt.show()
```
