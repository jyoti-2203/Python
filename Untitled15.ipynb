A Disarium Number is a number in which the sum of its digits powered with
their respective positions is equal to the number itself. The digits are indexed
starting from 1. For example, 175 is a Disarium number because:

```python
num = input('Enter the number: ')
result = 0
length = len(num)
for i in range(0,length):
    result = result + int(num[i])**(i+1)
if result == int(num):
    print('Disarium')
else:
    print('Not')
```

A Harshad Number (or Niven Number) is a number that is divisible by the sum
of its digits. If a number n can be divided evenly by the sum of its digits, it is
considered a Harshad Number.

```python
num = input('Enter the number: ')
#length = len(num)
result = 0
for i in num:
    result = result + int(i)
if int(num)%result==0:
    print('Harshad')
else:
    print('Not')
```

A Keith Number (or Repfigit Number) is a number that appears as a term in a
special sequence that starts with its digits. The sequence is generated by
summing the previous n terms (where n is the number of digits of the number),
and the process continues until either the number itself appears in the
sequence (in which case it is a Keith Number) or the sum exceeds the number
(in which case it is not a Keith Number).

```python
num = input('Enter the number: ')
digits = [int(d) for d in num]
length = len(digits)
while 1==1:
    next_term = sum(digits)

    if next_term==int(num):
        print('Its a keith number')
        break
    elif next_term>int(num):
        print('Its not a keith number')
        break

    digits.append(next_term)
    digits.pop(0)
```
