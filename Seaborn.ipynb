```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head(n = 10)
```

For the heart.csv dataset loaded into a variable called df, we could use the following code to create a scatter plot in Matplotlib:

```python
plt.scatter(data = df, x = 'age', y = 'cholestrol')
```

The following code would create a similar scatter plot in Seaborn:

```python
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
```

```python
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
plt.show()
```

You can vary the figure size using plt.figure() with the figsize parameter. We could add a title using plt.title(). We could also change the axis names (labels) using plt.xlabel() and plt.ylabel():

```python
#plt.figure(figsize = (7, 7))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(
    data = df,
    x = 'age',
    y = 'cholestrol',
    s = 10, marker = '^',
    c = 'r',
    alpha = 0.5
)
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

#Bar plot

```python
df_bar = pd.read_csv('/content/churn_data.csv')
df_bar
```

```python
df_bar.info()
```

```python
df_bar.describe()
```

```python
plt.figure(figsize = (5, 3))
sns.countplot(data = df_bar , x = 'PaymentMethod')
plt.xlabel("Payment Mode")
plt.ylabel("Count")
plt.title('Count of Payment Modes')
plt.grid(visible = True, axis = 'y')
plt.show()
```

```python
plt.figure(figsize = (8, 3))
sns.countplot(data = df_bar , x = 'PaymentMethod')
plt.xlabel("Payment Mode")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.grid(visible = True, axis = 'y')
plt.show()
```

```python
df_bar.groupby('PaymentMethod')
```

```python
df_bar.groupby('PaymentMethod')['MonthlyCharges'].agg(['mean','median'])
```

```python
plt.figure(figsize = (5, 3))
sns.barplot(data = df_bar , x = 'PaymentMethod', y= 'MonthlyCharges')
plt.xlabel("Payment Mode")
plt.ylabel("MonthlyCharges")
plt.xticks(rotation = 45)
plt.grid(visible = False, axis = 'y')
plt.show()
```

Above bar chart at the end it showing some black line i.e. it is error. How to remove it.

```python
plt.figure(figsize = (5, 3))
sns.barplot(data = df_bar , x = 'PaymentMethod', y= 'MonthlyCharges', errorbar = None)
plt.xlabel("Payment Mode")
plt.ylabel("MonthlyCharges")
plt.xticks(rotation = 45)
plt.grid(visible = False, axis = 'y')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.barplot(data = df_bar , x = 'PaymentMethod', y= 'MonthlyCharges', errorbar = None, estimator = 'median',
            order = sorted(df_bar['PaymentMethod'].unique()) )
plt.xlabel("Payment Mode")
plt.ylabel("MonthlyCharges")
plt.xticks(rotation = 45)
plt.grid(visible = False, axis = 'y')
plt.show()
```
